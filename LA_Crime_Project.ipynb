```python
df = crime.drop(labels = ['modus_operandi','weapon_code','weapon_description','crime_code_2','crime_code_3','crime_code_4','cross_street'], axis = 1)
```

```python
df['date_reported']= pd.to_datetime(df['date_reported'], format = '%Y-%m-%d')
df['date_occurred'] = pd.to_datetime(df['date_occurred'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
```

```python
plt.style.use('seaborn-v0_8-dark-palette')
plt.figure(figsize =(5,6))
plt.hist(df_age['victim_age'],edgecolor = 'black' , color = 'forestgreen', bins = 15)
plt.xlabel('Age')
plt.ylabel('Increased crime')
plt.title('Increased crime as per Age')
plt.tight_layout()
```

```python
df['date_occurred_no_time']=df['date_occurred'].dt.date
```

```python
crime_count_per_day = df.groupby(['date_occurred_no_time' ,'crime_description']).size()
```

```python
Top_20_crime=crime_count_per_day.groupby('crime_description').mean().sort_index().reset_index(name = 'Average_daily_count').sort_values(by = 'Average_daily_count',ascending = False).head(20)
```

```python
Top_20_crime
```

```python
plt.barh(Top_20_crime['crime_description'],Top_20_crime['Average_daily_count'])
```

```python
plt.figure(figsize= (12,8))
plt.subplot(2,2,1)
crime_plot=df.groupby(['year','month']).size().plot(color= 'green', linewidth=2)
crime_plot.set_xlabel('year - month',fontsize = 10)
crime_plot.set_title('Crime Over Time')
plt.tight_layout()


plt.subplot(2,2,2)
plt.hist(df_age['victim_age'],edgecolor = 'black' , color = 'darkorchid', bins = 15)
plt.xlabel('Age')
plt.ylabel('Increased crime')
plt.title('Increased crime as per Age')
plt.tight_layout()


plt.subplot(2,2,3)
plt.barh(Top_20_crime['crime_description'],Top_20_crime['Average_daily_count'])
plt.xlabel('crime_description')
plt.ylabel('Average_daily_count')
plt.title('Top_20_crime')
plt.tight_layout()

```
