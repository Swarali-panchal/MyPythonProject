```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
world = pd.read_csv('Data for python course/2018-2019+Happiness.csv')
airbnb = pd.read_csv('Data for python course/airbnb.csv')
salaries = pd.read_csv('Data for python course/salaries.csv')
sales = pd.read_csv('Data for python course/web_sales.csv')
eth = pd.read_csv('Data for python course/Ethereum+Historical+Data.csv')
```

```python
average_sales =[2000,4000,6000,8000,10000]
years=[2015,2016,2017,2018,2019]
```

```python
plt.plot(years,average_sales)
```

```python
plt.bar(years,average_sales)
```

```python
salaries.groupby('experience_level')['salary_in_usd'].mean().reset_index()
```

```python
avg_salary_experience = salaries.groupby('experience_level')['salary_in_usd'].mean().reset_index().sort_values(by ='salary_in_usd',ascending = False)
```

```python
plt.plot(avg_salary_experience['experience_level'],avg_salary_experience['salary_in_usd'])
```

```python
plt.bar(avg_salary_experience['experience_level'],avg_salary_experience['salary_in_usd'])
plt.xlabel('experience_level')
plt.ylabel('salary_in_usd')
plt.title('average_experience')
```
